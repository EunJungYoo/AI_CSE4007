```python
import random
import numpy as np
from math import exp,log
from matplotlib import pyplot as plt
```

```python
X = np.array([[0,0],[1,0],[0,1],[1,1]])
Y = np.array([0,0,0,1])
Z = np.array([])
```

```python
class logistic_regression_model():
    def __init__(self):
        self.w = np.random.normal(size = 2)
        self.b = np.random.normal(size = 1)
        
    def sigmoid(self,z):
        return 1/(1+exp(-z))
    
    def predict(self,x):
        z = np.inner(self.w,x) + self.b
        a = self.sigmoid(z)
        return a
    
model = logistic_regression_model()
```

```python
def train(X,Y,model,lr):
    dw0 = 0.0
    dw1 = 0.0
    db = 0.0
    m = len(X)
    cost = 0.0
    for x,y in zip(X,Y):
        a = model.predict(x)
        if y==1:
            cost -= log(a)
        else:
            cost -= log(1-a)
        dw0 += (a-y)*x[0]
        dw1 += (a-y)*x[1]
        db +=(a-y)
        
    cost /=m
    model.w[0] -= lr * dw0/m
    model.w[1] -= lr * dw1/m
    model.b -= lr*db/m
    
    return cost
    
```

```python
for epoch in range(10000):
    cost = train(X,Y,model,10)
    if epoch % 1==0:
        print(epoch,cost)
        Z = np.append(Z, np.array([cost]))
        
plt.plot(Z)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.ylim([0,1])
```
